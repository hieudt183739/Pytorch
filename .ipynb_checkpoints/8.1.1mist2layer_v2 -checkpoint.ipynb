<h2 id="Model">Neural Network Module and Training Function</h2> 


Define the neural network module or class, with two hidden Layers 


<img src="https://ibm.box.com/shared/static/5wtclahun0f70qlwkn2kxzh3amnbq4zg.png" width="200" alt="Neural Network Model">


```python
# Create the model class using sigmoid as the activation function

class Net(nn.Module):
    
    # Constructor
    def __init__(self, D_in, H1, H2, D_out):
        super(Net, self).__init__()
        self.linear1 = nn.Linear(D_in, H1)
        self.linear2 = nn.Linear(H1, H2)
        self.linear3 = nn.Linear(H2, D_out)
    
    # Prediction
    def forward(self,x):
        x = torch.sigmoid(self.linear1(x)) 
        x = torch.sigmoid(self.linear2(x))
        x = self.linear3(x)
        return x
```

Define the class with the Tanh activation function 


```python
# Create the model class using Tanh as a activation function

class NetTanh(nn.Module):
    
    # Constructor
    def __init__(self, D_in, H1, H2, D_out):
        super(NetTanh, self).__init__()
        self.linear1 = nn.Linear(D_in, H1)
        self.linear2 = nn.Linear(H1, H2)
        self.linear3 = nn.Linear(H2, D_out)
    
    # Prediction
    def forward(self, x):
        x = torch.tanh(self.linear1(x))
        x = torch.tanh(self.linear2(x))
        x = self.linear3(x)
        return x
```

Define the class for the Relu activation function 


```python
# Create the model class using Relu as a activation function

class NetRelu(nn.Module):
    
    # Constructor
    def __init__(self, D_in, H1, H2, D_out):
        super(NetRelu, self).__init__()
        self.linear1 = nn.Linear(D_in, H1)
        self.linear2 = nn.Linear(H1, H2)
        self.linear3 = nn.Linear(H2, D_out)
    
    # Prediction
    def forward(self, x):
        x = torch.relu(self.linear1(x))  
        x = torch.relu(self.linear2(x))
        x = self.linear3(x)
        return x
```

Define a function to  train the model, in this case the function returns a Python dictionary to store the training loss and accuracy on the validation data 


```python
# Train the model

def train(model, criterion, train_loader, validation_loader, optimizer, epochs=100):
    i = 0
    useful_stuff = {'training_loss': [], 'validation_accuracy': []}  
    
    for epoch in range(epochs):
        for i, (x, y) in enumerate(train_loader):
            optimizer.zero_grad()
            z = model(x.view(-1, 28 * 28))
            loss = criterion(z, y)
            loss.backward()
            optimizer.step()
            useful_stuff['training_loss'].append(loss.data.item())
        
        correct = 0
        for x, y in validation_loader:
            z = model(x.view(-1, 28 * 28))
            _, label = torch.max(z, 1)
            correct += (label == y).sum().item()
    
        accuracy = 100 * (correct / len(validation_dataset))
        useful_stuff['validation_accuracy'].append(accuracy)
    
    return useful_stuff
```

## Myself Work - Create a DNN with 2 hidden layers
Thử nghiệm dùng 3 hàm tác động Sigmoid, Tank, Relu  
Hàm dự đoán Decision Function là Softmax Function.   
Vì là dùng hàm Softmax nên net3 ko nhất thiết phải cho qua Softmax làm gì, chỉ cần so sánh các net3 với nhau là được rồi vì nó giá trị Softmax  đồng biến với net3  
Tập Valid chỉ dùng để kiểm tra độ chính xác của bộ w hiện tại, chứ ko có tác dụng để tìm hyper parameter

```python
import torch
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
import torchvision.datasets as dataset
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
torch.manual_seed(2)

# Create Dataset
train_dataset = dataset.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())
valid_dataset = dataset.MNIST(root='./data', download=True, transform=transforms.ToTensor())

##### Create Module
# Activate Funcion: Sigmoid
class NetSigmoid(nn.Module):
    def __init__(self, input_dim, H1, H2, output_dim):
        super(NetSigmoid, self).__init__()
        self.net1 = nn.Linear(input_dim, H1)
        self.net2 = nn.Linear(H1, H2)
        self.net3 = nn.Linear(H2, output_dim)
        
    def forward(self, x):
        z1 = torch.sigmoid(self.net1(x))
        z2 = torch.sigmoid(self.net2(z1))
        z3 = self.net3(z2)
        return z3

    
# Activate Funcion: Tank
class NetTank(nn.Module):
    def __init__(self, input_dim, H1, H2, output_dim):
        super(NetTank, self).__init__()
        self.net1 = nn.Linear(input_dim, H1)
        self.net2 = nn.Linear(H1, H2)
        self.net3 = nn.Linear(H2, output_dim)
        
    def forward(self, x):
        z1 = torch.tanh(self.net1(x))
        z2 = torch.tanh(self.net2(z1))
        z3 = self.net3(z2)
        return z3
    

# Activate Funcion: Relu
class NetRelu(nn.Module):
    def __init__(self, input_dim, H1, H2, output_dim):
        super(NetRelu, self).__init__()
        self.net1 = nn.Linear(input_dim, H1)
        self.net2 = nn.Linear(H1, H2)
        self.net3 = nn.Linear(H2, output_dim)
        
    def forward(self, x):
        z1 = torch.relu(self.net1(x))
        z2 = torch.relu(self.net2(z1))
        z3 = self.net3(z2)
        return z3

# Loss Function
criterion = nn.CrossEntropyLoss()
```

```python
trainloader = DataLoader(dataset=train_dataset, batch_size=2000, shuffle=True)
validloader = DataLoader(dataset=valid_dataset, batch_size=5000, shuffle=False)
```

```python
def training(model, epochs):
    usefull_stuff = {'loss': [], 'accuracy': []}
    
    for epoch in range(epochs):
        for x,y in trainloader:
            x = x.view(-1, 28*28)
            yhat = model(x)
            loss = criterion(yhat, y)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            usefull_stuff['loss'].append(loss.item())
        
        correct = 0
        for x,y in validloader:
            x = x.view(-1, 28*28)
            yhat = model(x)
            _,label = torch.max(yhat, 1)
            correct += (label==y).sum().item()
        usefull_stuff['accuracy'].append(correct/len(valid_dataset))
        
    return usefull_stuff
            
```

```python
input_dim = 28*28
H1 = 50
H2 = 50
output_dim = 10
epochs = 30
learing_rate = 0.01
```

```python
# Model NetSigmoid
model = NetSigmoid(input_dim, H1, H2, output_dim)
optimizer = optim.SGD(model.parameters(), lr=learing_rate)
training_result_sigmoid = training(model, epochs)
```

```python
# Model NetSigmoid
model = NetTank(input_dim, H1, H2, output_dim)
optimizer = optim.SGD(model.parameters(), lr=learing_rate)
training_result_tank = training(model, epochs)
```

```python
# Model NetSigmoid
model = NetRelu(input_dim, H1, H2, output_dim)
optimizer = optim.SGD(model.parameters(), lr=learing_rate)
training_result_relu = training(model, epochs)
```

```python
plt.plot(training_result_tank['accuracy'], label = 'tanh')
plt.plot(training_result_sigmoid['accuracy'], label = 'sigmoid')
plt.plot(training_result_relu['accuracy'], label = 'relu') 
plt.ylabel('validation accuracy')
plt.xlabel('Epoch')   
plt.legend()
plt.show()
```

```python
plt.plot(training_result_tank['loss'], label = 'tanh')
plt.plot(training_result_sigmoid['loss'], label = 'sigmoid')
plt.plot(training_result_relu['loss'], label = 'relu') 
plt.ylabel('Loss')
plt.xlabel('Iteration')   
plt.legend()
plt.show()
```
