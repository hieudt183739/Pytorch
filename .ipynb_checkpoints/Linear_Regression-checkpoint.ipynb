```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

## Khởi tạo bộ dữ liệu

```python

X = torch.arange(-3, 3, 0.1).view(-1, 1)    # chuyen 2D tensor 1 cot, chuyen ve de su dung tinh ma tran cho nhanh
Y = X * 5
Y += 0.2 * torch.randn(X.size()) # add some noise
plt.scatter(X, Y, color='red', s=20)
```

## Hàm f, loss function, class để theo dõi sự thay đổi của w

```python
w = torch.tensor(10.0, requires_grad=True)
lr = 0.1    # learning rate

def forward(x):
    return w * x

# loss function
def critrion(yhat, y):
    return torch.mean((yhat - y) ** 2)

class keep_track_w:
    
    # Chuoi gia tri cua w va loss voi cac gia tri w do
    def __init__(self, start, stop):
        tmp = w.data
        self.w_values = torch.arange(start, stop, 0.1)
        self.loss_values = [critrion(forward(X), Y).detach() for w.data in self.w_values]
        w.data = tmp
    
    def __call__(self, w, loss, epoch):
        plt.figure(figsize=(15, 6))

        # plot theo doi w
        plt.subplot(121)
        plt.plot(self.w_values, self.loss_values, color='b')    # duong cong loss
        plt.scatter(w, loss, color='red')                       # w hien tai 
        
        # plot ket qua duong thang nhan dc
        plt.subplot(122)
        plt.scatter(X, Y, color='red', s=20)                    
        plt.plot(X, forward(X).detach().numpy())                # dg thang voi w hien tai
        
        plt.title(f'Epoch {epoch}')
        plt.figure()
    
k = keep_track_w(-5, 15)
```

## Linear Regression Module

```python

def train_module(iter):
    for epoch in range(iter):
        Yhat = forward(X)
        loss = critrion(Yhat, Y)
        
        # visualization value of w
        k(w.data, loss.data, epoch)
        
        # tinh delta w
        loss.backward()
        w.data = w.data - lr * w.grad.data
        w.grad.zero_()    # dat grad ve 0, neu ko dat thi delta w se cong dong voi delta w truoc do
        
    else:
        Yhat = forward(X)
        loss = critrion(Yhat, Y)        
        k(w.data, loss.data, 4)   
        
      
```

```python
train_module(4)
```

## Optimization in Pytorch

```python
import torch
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, Dataset
from torch import nn, optim
```

### 1. Tạo Dataset Class

```python
class Data(Dataset):
    
    # Contructor
    def __init__(self):
        self.X = torch.arange(-10, 10, 0.1).view(-1,1)
        self.Y = -3 * self.X + 0.1 * torch.randn(self.X.size())
        self.length = self.X.shape[0]    # so hang
        # print(self.Y)
    
    # Getter
    def __getitem__(self, index):
        return self.X[index], self.Y[index]
    
    # Get Length
    def __len__(self):
        return self.length
```

### 2. Tạo Linear Regression Class

```python
class Linear_Regressin(nn.Module):
    
    # Contructor
    def __init__(self, input_size, output_size):
        super(Linear_Regressin, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
    
    # Prediction
    def forward(self, x):
        yhat = self.linear(x)
        return yhat
    
# Choose Loss function
criteration = nn.MSELoss()    
```

### 3. Tạo dataset và model

```python
dataset = Data()
model = Linear_Regressin(1, 1)
```

### 4. Chọn phương pháp cập nhập trọng só

```python
# Choose w update method
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

### 5. Tạo DataLoader Object

```python
trainloader = DataLoader(dataset=dataset, batch_size=10)
```

### 6. Tạo Train Module

```python
def train_module_BGD(epochs):
    for epoch in range(epochs):
        for x,y in trainloader:
            yhat = model(x)
            # print(y, yhat)
            loss = criteration(yhat, y)
            optimizer.zero_grad()    # cho w.grad và bias.grad về 0 để nó không bị dồn các delta w,bias
            loss.backward()
            optimizer.step()    # cập nhập w, bias
            
```

### 7. Training

```python
# khởi tạo bộ trọng số 
model.state_dict()['linear.weight'][0] = -15
model.state_dict()['linear.bias'][0] = -10

# training
train_module_BGD(10)
```
