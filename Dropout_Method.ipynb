```python
# Import the libraries we need for this lab

import torch
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from matplotlib.colors import ListedColormap
from torch.utils.data import Dataset, DataLoader
```

Use this function only for plotting:


```python
# The function for plotting the diagram

def plot_decision_regions_3class(data_set, model=None):
    cmap_light = ListedColormap([ '#0000FF','#FF0000'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#00AAFF'])
    X = data_set.x.numpy()
    y = data_set.y.numpy()
    h = .02
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1 
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1 
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    newdata = np.c_[xx.ravel(), yy.ravel()]
    
    Z = data_set.multi_dim_poly(newdata).flatten()
    f = np.zeros(Z.shape)
    f[Z > 0] = 1
    f = f.reshape(xx.shape)
    if model != None:
        model.eval()
        XX = torch.Tensor(newdata)
        _, yhat = torch.max(model(XX), 1)
        yhat = yhat.numpy().reshape(xx.shape)
        plt.pcolormesh(xx, yy, yhat, cmap=cmap_light)
        plt.contour(xx, yy, f, cmap=plt.cm.Paired)
    else:
        plt.contour(xx, yy, f, cmap=plt.cm.Paired)
        plt.pcolormesh(xx, yy, f, cmap=cmap_light) 

    plt.title("decision region vs True decision boundary")
```

Use this function to calculate accuracy: 


```python
# The function for calculating accuracy

def accuracy(model, data_set):
    _, yhat = torch.max(model(data_set.x), 1)
    return (yhat == data_set.y).numpy().mean()
```

<h2 id="Model_Cost">Create the Model, Optimizer, and Total Loss Function (Cost)</h2>


Create a custom module with three layers. <code>in_size</code> is the size of the input features, <code>n_hidden</code> is the size of the layers, and <code>out_size</code> is the size. <code>p</code> is the dropout probability. The default is 0, that is, no dropout.


```python
# Create Net Class

class Net(nn.Module):
    
    # Constructor
    def __init__(self, in_size, n_hidden, out_size, p=0):
        super(Net, self).__init__()
        self.drop = nn.Dropout(p=p)
        self.linear1 = nn.Linear(in_size, n_hidden)
        self.linear2 = nn.Linear(n_hidden, n_hidden)
        self.linear3 = nn.Linear(n_hidden, out_size)
    
    # Prediction function
    def forward(self, x):
        x = F.relu(self.drop(self.linear1(x)))
        x = F.relu(self.drop(self.linear2(x)))
        x = self.linear3(x)
        return x
```

Create two model objects: <code>model</code> had no dropout and <code>model_drop</code> has a dropout probability of 0.5:


```python
# Create two model objects: model without dropout and model with dropout

model = Net(2, 300, 2)
model_drop = Net(2, 300, 2, p=0.5)
```

<h2 id="BGD">Train the Model via Mini-Batch Gradient Descent</h2>


Set the model using dropout to training mode; this is the default mode, but it's  good practice to write this in your code : 


```python
# Set the model to training mode

model_drop.train()
```

Train the model by using the Adam optimizer. See the unit on other optimizers. Use the Cross Entropy Loss:


```python
# Set optimizer functions and criterion functions

optimizer_ofit = torch.optim.Adam(model.parameters(), lr=0.01)
optimizer_drop = torch.optim.Adam(model_drop.parameters(), lr=0.01)
criterion = torch.nn.CrossEntropyLoss()
```

Initialize a dictionary that stores the training and validation loss for each model:


```python
# Initialize the LOSS dictionary to store the loss

LOSS = {}
LOSS['training data no dropout'] = []
LOSS['validation data no dropout'] = []
LOSS['training data dropout'] = []
LOSS['validation data dropout'] = []
```

Run 500 iterations of batch gradient gradient descent: 


```python
# Train the model

epochs = 500

def train_model(epochs):
    
    for epoch in range(epochs):
        #all the samples are used for training 
        yhat = model(data_set.x)
        yhat_drop = model_drop(data_set.x)
        loss = criterion(yhat, data_set.y)
        loss_drop = criterion(yhat_drop, data_set.y)

        #store the loss for both the training and validation data for both models 
        LOSS['training data no dropout'].append(loss.item())
        LOSS['validation data no dropout'].append(criterion(model(validation_set.x), validation_set.y).item())
        LOSS['training data dropout'].append(loss_drop.item())
        model_drop.eval()
        LOSS['validation data dropout'].append(criterion(model_drop(validation_set.x), validation_set.y).item())
        model_drop.train()

        optimizer_ofit.zero_grad()
        optimizer_drop.zero_grad()
        loss.backward()
        loss_drop.backward()
        optimizer_ofit.step()
        optimizer_drop.step()
        
train_model(epochs)
```

Set the model with dropout to evaluation mode: 


```python
# Set the model to evaluation model

model_drop.eval()
```

Test the model without dropout on the validation data: 


```python
# Print out the accuracy of the model without dropout

print("The accuracy of the model without dropout: ", accuracy(model, validation_set))
```

Test the model with dropout on the validation data: 


```python
# Print out the accuracy of the model with dropout

print("The accuracy of the model with dropout: ", accuracy(model_drop, validation_set))
```

You see that the model with dropout performs better on the validation data.

```python
class Data(Dataset):
    def __init__(self, N_sample=1000, train=True):
        self.x = np.zeros((N_sample, 2))    # 2D (x1,x2)
        self.y = np.zeros((N_sample, 1))    # 2D label, 1 column
        
        self.x[:,0] = np.random.uniform(0, 1, size=N_sample)    # x1
        self.x[:,1] = np.random.uniform(0, 1, size=N_sample)    # x2
        
        self.f = 1.6 * (self.x[:,0]**2) -1.2*self.x[:,0] + 0.6
        self.y[self.x[:,1] > self.f] = 1
        
        # to nhiu cho b d liu train, tc l gn sai nhn cho 1 vi sample ngu nhin
        if train == True:
            np.random.seed(1)
            indicate1 = np.random.randint(0, 1000, size=int(0.3*N_sample), dtype=int)
            indicate2 = np.random.randint(0, 1000, size=int(0.3*N_sample), dtype=int)
            for i,j in zip(indicate1, indicate2): 
                self.y[i,:] = 0
                self.y[j,:] = 1
          
        self.x = torch.from_numpy(self.x).type(torch.FloatTensor)
        self.y = torch.from_numpy(self.y).type(torch.LongTensor)
    
    def __getitem__(self, index):
        return self.x[index,:], self.y[index]
    
    def __len__(self):
        return self.x.shape[0]
    
    def plot_data(self):
        x1 = self.x[:,0].numpy()
        x2 = self.x[:,1].numpy()
        label = self.y.view(-1).numpy()
        f = self.f
        
        # biu din d liu trn mt phng 2D
        plt.plot(x1[label==1],x2[label==1], 'ro', label='label=1')
        plt.plot(x1[label==0],x2[label==0], 'bo', label='label=0')
        plt.xlabel('x1')
        plt.ylabel('x2', rotation=0)
         
        # biu din ng phn chia d liu thc t
        # cn sp xp li (x1, f) vi th t x1 tng dn th n mi c dng l g thng
        indicate = np.argsort(x1)[::-1]    # v tr cc gi tr t b ti ln    
        x1 = x1[indicate]
        f = f[indicate]
        plt.plot(x1, f, color='black', label='True Decision Boundary')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # t ch gii bn ngoi plot
        plt.show()
    
```

```python
train_dataset = Data(train=True)
train_dataset.plot_data()
```

```python
valid_dataset = Data(train=False)
valid_dataset.plot_data()
```

## Create Model


```python
class NN(nn.Module):
    def __init__(self, input_size, H, output_size, p=0):
        super(NN, self).__init__()
        self.drop = nn.Dropout(p=p)
        self.linear1 = nn.Linear(input_size, H)
        self.linear2 = nn.Linear(H, H)
        self.linear3 = nn.Linear(H, H)
        self.linear4 = nn.Linear(H, H)
        self.linear5 = nn.Linear(H, output_size)
        
    def forward(self, x):
        x = torch.relu(self.drop(self.linear1(x)))
        x = torch.relu(self.drop(self.linear2(x)))
        x = torch.relu(self.drop(self.linear3(x)))
        x = torch.relu(self.drop(self.linear4(x)))
        z = self.linear5(x)
        return z

criterion = nn.CrossEntropyLoss()

model = NN(2, 50, 2)        # Network not using Dropout
model_drop = NN(2, 50, 2, 0.2)   # Network using Dropout
model_drop.train()
```

## Create DataLoader, Optimizer for Non-Dropout Network

```python
trainloader = DataLoader(dataset=train_dataset, batch_size=100)

optimazer = optim.Adam(model.parameters(), lr=0.01)              # for NN not using Dropout
optimazer_drop = optim.Adam(model_drop.parameters(), lr=0.01)    # for NN using Dropout
```

## Train

```python
COST = {'non_drop': [], 'drop': []}
ACC = {'non_drop': [], 'drop': []}
def train(epochs):
    for epoch in range(epochs):
        total = 0
        total_drop = 0
        for x,y in trainloader:
            yhat = model(x)
            yhat_drop = model_drop(x)
            loss = criterion(yhat, y.view(-1))
            loss_drop = criterion(yhat_drop, y.view(-1))
            
            # COST['non_drop'].append(loss.item())
            # COST['drop'].append(loss_drop.item())
            total += loss.item()
            total_drop += loss_drop.item()
            
            optimazer.zero_grad()
            optimazer_drop.zero_grad()
            loss.backward()
            loss_drop.backward()
            optimazer.step()
            optimazer_drop.step()
        
        COST['non_drop'].append(loss.item() / len(train_dataset))
        COST['drop'].append(loss_drop.item()/ len(train_dataset)) 
        
        # Test Accuracy
        x = valid_dataset[:][0]
        y = valid_dataset[:][1].view(-1)
        z = model(x)
        _,yhat = torch.max(z, 1)   # hm torch.max() lun tr v 1D Tensor
        # print(z, yhat)
        correct = (y==yhat).sum().item()
        ACC['non_drop'].append(correct / len(valid_dataset))
        
        model_drop.eval()
        z_drop = model_drop(x)
        _,yhat_drop = torch.max(z_drop, 1)
        correct_drop = (y==yhat_drop).sum().item()
        ACC['drop'].append(correct_drop / len(valid_dataset))
        model_drop.train()
            
            
```

```python
train(500)
```

## View Result

```python
plt.plot(np.log(COST['non_drop']), label='non')
plt.plot(np.log(COST['drop']), label='drop')
plt.legend()
plt.show()

plt.plot(ACC['non_drop'])
plt.plot(ACC['drop'])
plt.show()
```
